```python
#PRAKTIKUM 1

class BankAccount:
    def __init__(self, owner, balance):

        self.__owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} telah ditambahkan ke akun {self.__owner}.")
        else:
            print("Jumlah deposit harus lebih dari 0.")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} telah ditaril dari akun {self.__owner}.")
        else:
            print("Saldo tidak mencukupi.")

    def get_balance(self):
        return self.__balance
    
if __name__ == "__main__":
    
    alice_account = BankAccount(owner="Alice", balance=1000)

    alice_account.deposit(500) #berhasil
    alice_account.deposit(-100) #gagal

    alice_account.withdraw(300) #berhasil
    alice_account.withdraw(2000) #gagal

    #get saldo
    current_balance = alice_account.get_balance()
    print(f"Saldo terakhir di akun {alice_account._BankAccount__owner}: {current_balance}")
```

```python
#PRAKTIKUM 2

class Employee:
    def __init__(self, name, salary):

        #attr privat dengan double underscore
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase):
        """"method untuk update gaji dengan validasi"""
        if increase > 0:
            self.__salary += increase
            print(f"Gaji telah dinaikkan sebesar {increase}.")
        else:
            print("Nilai kenaikan harus lebih dari 0.")

    def set_salary(self, new_salary):
        if new_salary >= 0:
            self.__salary = new_salary
            print(f"Gaji diatur ulang menjadi {new_salary}.")
        else:
            print("Gaji tidak dapat bernilai negatif.")

    def get_salary(self):
        """"method untuk get info gaji"""
        return self.__salary
    
    def get_employee_info(self):
        return f"Employee: {self.__name}, Gaji: {self.__salary}"
    
#Contoh Penggunaan
if __name__ == "__main__":

    #membuat objek
    employee1 = Employee("Jhon Doe", 50000)
    print(employee1.get_employee_info())

    #update gaji
    employee1.update_salary(5000)
    print(f"Gaji setelah kenaikan: {employee1.get_salary()}")

    #atur ulang nilai gaji
    employee1.set_salary(60000)
    print(f"Gaji setelah kenaikan: {employee1.get_employee_info()}")
```

```python
#PRAKTIKUM 3

class SimpleExample:
    def __init__(self, name):
        """
        Konstruktor: Dipanggil saat objek dibuat.
        Menyimpan nilai 'name' dan mencetak pesan pembuatan objek.
        """
        self.name = name
        print(f"Kontruktor: Objek '{self.name}' telah diisi / dibuat.")

    def __del__(self):
        """
        Destruktor: Dipanggil saat objek dihapus.
        Mencetak pesan bahwa objek sedang dihapus.
        """
        print(f"DestruktorL Objek '{self.name}' sedang dihapus.")

def main():
    print("Program dimulai. \n")

    #membuat objek SimpleExample
    obj = SimpleExample("Demo")
    print("Program sedang berjalan ... \n")

    #menghapus objek scr eksplisit
    del obj
    print("Objek telah dihapus secara eksplisit. \n")

    print("Program selesai.")

if __name__ == "__main__":
    main()
```

```python
#PRAKTIKUM 4

class FileLogger:
    def __init__(self, filename):
        """
        Konstruktor: Membuka file log untuk menulis pesan.
        :param filename: Nama file tempat pesan log akan ditulis.
        """
        self.filename = filename
        try:
            self.file = open(filename, "a")  # Membuka file dalam mode append
            print(f"File log '{self.filename}' dibuka untuk logging.")
        except:
            print(f"File '{self.filename}' gagal dibuka.")
            self.file = None

    def write_log(self, message):
        """
        Menulis pesan log ke dalam file.
        :param message: Pesan yang akan ditulis ke file.
        """
        if self.file:
            self.file.write(message + "\n")
            print(f"Pesan log ditulis: {message}")
        else:
            print("Tidak dapat menulis log karena file tidak tersedia.")

    def __del__(self):
        """
        Destruktor: Menutup file log ketika objek dihapus.
        """
        if hasattr(self, 'file') and self.file and not self.file.closed:
            self.file.close()
            print(f"File log '{self.filename}' telah ditutup.")

# Contoh penggunaan kelas FileLogger
logger = FileLogger("application.log")
logger.write_log("Aplikasi dimulai.")
logger.write_log("Menjalankan proses A...")
logger.write_log("Proses A selesai.")
logger.write_log("Aplikasi dihentikan.")

# Menghapus objek logger secara eksplisit
del logger

```

```python
#PRAKTIKUM 5

class Calculator:
    def __init__(self, initial_value=0):
        """
        Konstruktor untuk kelas Calculator.
        :param initial_value: Nilai awal kalkulator yang diinisialisasi.
        """
        self.value = initial_value
        print(f"Calculator diinisialisasi dengan nilai awal {self.value}")

    def add(self, number):
        """
        Menambahkan angka ke nilai yang tersimpan di objek.
        :param number: Angka yang akan ditambahkan.
        """
        self.value += number
        print(f"Nilai setelah ditambah {number} adalah: {self.value}")

    def subtract(self, number):
        """
        Mengurangi angka dari nilai yang tersimpan di objek.
        :param number: Angka yang akan dikurangkan.
        """
        self.value -= number
        print(f"Nilai setelah dikurangi {number} adalah: {self.value}")

    def reset(self):
        """
        Mereset nilai kalkulator ke 0.
        """
        self.value = 0
        print("Nilai kalkulator telah direset ke 0.")

    def show_value(self):
        """
        Menampilkan nilai saat ini dari kalkulator.
        """
        print(f"Nilai saat ini adalah: {self.value}")

    def __del__(self):
        print("Calculator telah dihapus dari memori.")

# Membuat objek calculator dengan nilai awal 0
calc1 = Calculator()
calc1.add(10)
calc1.subtract(3)
calc1.show_value()
calc1.reset()
calc1.show_value()

# Membuat objek calculator lain dengan nilai awal 50
calc2 = Calculator(50)
calc2.subtract(20)
calc2.show_value()

# Menghapus objek
del calc1
del calc2
```

```python
#PRAKTIKUM 6

class Person:
    def __init__(self, name, age):
        """
        Konstruktor untuk menginisialisasi objek Person dengan nama dan usia.
        """
        self.__name = name
        self.__age = 0  # nilai awal sebelum setter dijalankan
        self.age = age  # akan divalidasi lewat setter

    # Getter dan Setter untuk name
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if value.strip():
            self.__name = value
        else:
            print("Nama tidak boleh kosong.")

    # Getter dan Setter untuk age
    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if 0 <= value <= 120:
            self.__age = value
        else:
            print("Usia tidak valid (harus antara 0 hingga 120).")

    def show_info(self):
        print(f"Nama: {self.name}")
        print(f"Usia: {self.age}")


# Contoh penggunaan
p1 = Person("Siti", 25)
p1.show_info()

# Mengubah nilai dengan validasi
p1.name = "Ayu"
p1.age = 30
p1.show_info()

# Uji nilai yang tidak valid
p1.name = ""
p1.age = 200

# Output final
p1.show_info()
```

```python
#PENUGASAN

class Student:
    def __init__(self, name, score):
        self.__name = name
        self.__score = 0
        self.__grade = 'E'
        self.score = score #setter langsung dipakai untuk validasi dan update grade

    #Getter dan setter untuk name
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, value):
        if value.strip():
            self.__name = value
        else:
            print("Nama tidak boleh kosong!")

    #Getter dan setter untuk score
    @property
    def score(self):
        return self.__score
    
    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()
        else:
            print("Score harus di antara 0 dan 100!")

    #Getter untuk grade
    @property
    def grade(self):
        return self.__grade
    
    #method internal untuk update grade
    def __update_grade(self):
        if self.__score >= 90:
            self.__grade = 'A'
        if self.__score >= 80:
            self.__grade = 'B'
        if self.__score >= 70:
            self.__grade = 'C'
        if self.__score >= 60:
            self.__grade = 'D'
        else:
            self.__grade = 'E'

    #method untuk menampilkan info mahasiswa
    def show_info(self):
        print(f"Nama Mahasiswa: {self.name}")
        print(f"Nilai: {self.score}")
        print(f"Grade: {self.grade}")

    #Destructor
    def __del__(self):
        print(f"Data Mahasiswa {self.__name} telah dihapus dari sistem.")

# Contoh penggunaan kelas
s1 = Student("Siti", 87)
s1.show_info()

print("\nNilai diubah ... ")
s1.score = 93
s1.show_info()

#hapus objek
del s1
```
